## Estad√≠sticas de la NBA para la temporada 2019-2020
### Los datos presentados se muestran hasta 19-02-2020

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
path_archivos = "C://Users//Andres//Desktop//PYTHON//py-pantoja-pino-andres-sebastian//Proyecto_2B//archivos//"

columnas = ['Equipo', 'Porcentajes']

df = pd.read_csv(path_archivos + 'total_standings.csv', usecols = columnas);
df = df.sort_values(by ='Porcentajes' ,ascending=False)
fig, ax = plt.subplots()
ax.bar(
       df['Equipo'],
       df['Porcentajes'])
plt.xticks(rotation=90)
fig.suptitle('Win - Loss Ratio ', fontsize=20)

plt.show()
```

```python
standings_columns = ['Equipo', 'Victorias', 'Derrotas']

df = pd.read_csv(path_archivos + 'standings_east.csv', usecols = standings_columns);
df = df.sort_values(by ='Victorias' ,ascending=False)
ratio = df['Victorias'] / df['Derrotas']
fig, ax = plt.subplots()
ax.bar(
       df['Equipo'],
       ratio)
plt.xticks(rotation=90)
fig.suptitle('Standings in the East ', fontsize=20)

plt.show()
```

```python
standings_columns = ['Equipo', 'Victorias', 'Derrotas']

df = pd.read_csv(path_archivos + 'standings_west.csv', usecols = standings_columns);
df = df.sort_values(by ='Victorias' ,ascending=False)
ratio = df['Victorias'] / df['Derrotas']
fig, ax = plt.subplots()
ax.bar(
       df['Equipo'],
       ratio)
plt.xticks(rotation=90)
fig.suptitle('Standings in the West ', fontsize=20)

plt.show()
```

```python
playoff_prob_columns = ['Equipo', 'Probabilidad']
df = pd.read_csv(path_archivos + 'win_probs.csv', usecols = playoff_prob_columns);
df = df.sort_values(by ='Probabilidad' ,ascending=False)
df = df[df['Probabilidad']>5]
colors = ['yellowgreen', 'gold']
fig, ax1 = plt.subplots()
ax1.pie(
    df['Probabilidad'], 
    labels = df['Equipo'],
    explode = [0.05] * len(df['Equipo']),
    autopct = '%1.2f%%', 
    shadow = True, 
    radius = 1,
    colors = colors,
    startangle = 0)
fig.suptitle('Chance of Winning the Title', fontsize=20)
fig.figsize = (5,5)
plt.show()
```

```python
mvp_prob_columns = ['Jugador', 'Probabilidad']
df = pd.read_csv(path_archivos + 'mvp_probs.csv', usecols = mvp_prob_columns);
df = df[df['Probabilidad']>5]
colors = ['yellowgreen', 'gold']
fig, ax1 = plt.subplots()
ax1.pie(
    df['Probabilidad'], 
    labels = df['Jugador'],
    explode = [0.05] * len(df['Jugador']),
    autopct = '%1.2f%%', 
    shadow = True, 
    radius = 1,
    colors = colors,
    startangle = 0)
fig.suptitle('Chance of Winning MVP', fontsize=20)
fig.figsize = (5,5)
plt.show()
```

```python
from math import pi
stats_columns = ['TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']

df = pd.read_csv(path_archivos + 'mvp_stats.csv', usecols = stats_columns);
AttNo = len(stats_columns)
datos_2019 = df.iloc[0].tolist()
datos_2019 +=  datos_2019[:1]

datos_2020 = df.iloc[1].tolist()
datos_2020 += datos_2020[:1]

angles = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles += angles [:1]

angles2 = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles2 += angles2 [:1]

ax = plt.subplot(111, polar=True)

plt.xticks(angles,stats_columns)

ax.plot(angles,datos_2019)
ax.fill(angles, datos_2019, 'teal', alpha=0.1)


ax.plot(angles2, datos_2020)
ax.fill(angles2, datos_2020, 'red', alpha=0.1)

#Give the plot a title and show it
plt.figtext(0.2,0.9,"Projected MVP 2020",color="red")
plt.figtext(0.2,0.85,"v")
plt.figtext(0.2,0.8,"MVP 2019",color="teal")
plt.show()
plt.show()
```
